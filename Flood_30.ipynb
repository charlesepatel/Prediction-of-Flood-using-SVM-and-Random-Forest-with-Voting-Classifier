```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Subset the DataFrame with the selected columns
subset_df = data[columns_of_interest]

# Basic statistics
statistics = subset_df.describe()

# Histograms for each column
subset_df.hist(figsize=(20, 15), bins=20)
plt.suptitle('Histograms of the Columns', fontsize=16)
plt.show()

```

```python
# Box plots for each column
plt.figure(figsize=(20, 10))
sns.boxplot(data=subset_df)
plt.xticks(rotation=70)
plt.title('Box Plots of the Columns', fontsize=16)
plt.show()

```

```python
# Calculate Q1, Q3, and IQR
Q1 = subset_df.quantile(0.25)
Q3 = subset_df.quantile(0.75)
IQR = Q3 - Q1

# Define the range for outlier detection (e.g., 1.5 times IQR)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the DataFrame
subset_df_no_outliers = subset_df[(subset_df >= lower_bound) & (subset_df <= upper_bound)]

# Display the DataFrame without outliers
subset_df_no_outliers
```

```python
subset_df_no_outliers.describe()
```

```python
subset_df_no_outliers.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

# Define imputation strategy (median or mean)
imputation_strategy = 'mean'  # or 'mean'

# Create a SimpleImputer object with the chosen strategy
imputer = SimpleImputer(strategy=imputation_strategy)

# Iterate over each column in the DataFrame
for column in subset_df_no_outliers.columns:
    # Skip the 'FloodProbability' column
    if column == 'FloodProbability':
        continue
    
    # Impute missing values in the current column
    subset_df_no_outliers[column] = imputer.fit_transform(subset_df_no_outliers[[column]])

# Check if there are any remaining missing values after imputation
print(subset_df_no_outliers.isnull().sum())
```
